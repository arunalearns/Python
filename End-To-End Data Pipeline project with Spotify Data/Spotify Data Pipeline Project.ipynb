```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
#add your keys here
client_credentials_manager = SpotifyClientCredentials(client_id="", client_secret="")
```

```python
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
data = sp.playlist_tracks('3cEYpjA9oz9GiPac4AsH4n')
```

```python
album_list = list()
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    album_element = {'album_id':album_id,
                     'name':album_name,
                     'release_date':album_release_date,
                     'total_tracks':album_total_tracks,
                     'url':album_url}
    #print(album_element)  # Creating the data from json (messy data) to dictionary album element is a dictinary
    album_list.append(album_element)  # Creating list of elements
```

```python
# Method 1
artist_list=list()
for row in data['items']:
         artist_id = row['track']['artists'][0]['id']
         artist_name = row['track']['artists'][0]['name']
         external_url = row['track']['artists'][0]['href']
         artist_element = {
                     'artist_id':artist_id,
                     'artist_name':artist_name,
                     'external_url':external_url
                     }
         artist_list.append(artist_element)

    
```

```python
artist_list
```

```python
# Method 2
#data['items']
artist_list=list()
for row in data['items']:
    for key,value in row.items():
         if key == 'track':
            # print(value['artists'])
             for artist in value['artists']:
                 #print(artist['id'])
                 artist_dict = {
                     'artist_id' : artist['id'],
                     'artist_name' : artist['name'],
                     'external_urls' : artist['external_urls'],
                 }
                 artist_list.append(artist_dict)   
                 
                 
```

```python
song_list = list()
for row in data['items']:
    for key, value in row.items():  # because it's a dictionary
        if key == 'track':
            track = value  # store the track value in a separate variable to avoid overwriting 'value'
            song_element = {
                'song_id': track['id'],
                'song_name': track['name'],
                'duration_ms': track['duration_ms'],
                'url': track['external_urls']['spotify'],
                'popularity': track['popularity'],
                'song_added': row['added_at'],  # Added 'added_at' here
                'album_id': track['album']['id'],
                'artist_id': track['album']['artists'][0]['id']
            }
            song_list.append(song_element)
```

```python
song_list
```

```python
# ALBUM Dataframe
album_df = pd.DataFrame.from_dict(album_list)
album_df = album_df.drop_duplicates(subset=['album_id']) # Removing Duplicates
album_df
```

```python
# ARTIST Dataframe
artist_df = pd.DataFrame.from_dict(artist_list)
artist_df = artist_df.drop_duplicates(subset=['artist_id']) # Removing Duplicates
artist_df
```

```python
#Song Dataframe
song_df = pd.DataFrame.from_dict(song_list)
song_df
```
